# Вариант 5

```python
from random import randint
import numpy as np

class FibNumberGenerator:
    mod = 0xFFFFFFF
    numbers = [0, 1, 1]

    def get_number(self, i: int):
        
        while len(self.numbers) <= i:
            index = len(self.numbers)

            self.numbers.append(self.numbers[index - 1] + self.numbers[index - 2])

        return self.numbers[i] % self.mod

gen = FibNumberGenerator()
```

## Задание 1

```python
size = (512, 128)
generation_range = (-8, 8)

matrix = np.matrix([[randint(a=generation_range[0], b=generation_range[1]) for _ in range(size[1])] for _ in range(size[0])])

print(*matrix, sep='\n')
```

## Задание 2

```python
size = 128
first = 0
step = 3

replace = 128

vector = first + np.arange(0, size) * step

matrix[replace] = vector

print('vector')
print(vector)
print('matrix')
print(*matrix, sep='\n')
```

## Задание 3

```python
size = 512
replace = 100

column = [gen.get_number(i + 1) for i in np.arange(0, size)]

matrix[:, replace] = np.array([np.array([x], dtype=str) for x in column])

print('vector')
print(column)
print('matrix')
print(*matrix, sep='\n')
```
