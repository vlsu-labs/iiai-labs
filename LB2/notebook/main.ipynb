# Вариант 5

```python
import math
from matplotlib import pyplot as plt
from matplotlib.pyplot import subplots_adjust
import numpy as np

def r(a, d):
    if a == 0:
        return None

    if round(d, 8) == 0:
        return 0
    if d < 0:
        return None
    
    return a*math.sqrt(d)
```

## Задание 1

```python
a = 5


step = (2 - -2) / 20
x = np.arange(-2, 2 + step, step)

y = [r(a, i) for i in x]

fig = plt.figure(num=1, facecolor='w', edgecolor='k')
fig.set_figwidth(5)
fig.set_figheight(5)
# subplots_adjust(hspace=0.4)

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, color='red', label='график',marker='o', linestyle='dotted')
ax.set_title('Вариант 5')
ax.set_xlabel('ось x')
ax.set_ylabel('ось y')
ax.grid()
ax.legend()
# plt.yticks(np.arange(-1, max([i if i is not None else 0 for i in y])+1, 0.5))

plt.show()

```

## Задание 5

```python

def z(x, y):
    return 4 * np.sin(math.pi*2) * np.cos(math.pi*1.5*y)*(1 - x**2) * y * (1 - y) 

step = 0.05
xx = np.arange(-1, 1 + step, step)
yy = np.arange(0, 1 + step, step)

X, Y = np.meshgrid(xx, yy)

Z = z(X, Y)

fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot_wireframe(X, Y, Z, color ='green')
ax.set_title('Вариант 5');
```
