# Вариант 5

```python
import math
from matplotlib import pyplot as plt
from matplotlib.pyplot import subplots_adjust
import numpy as np

def r(a, d):
    if a == 0:
        return None

    if round(d, 8) == 0:
        return 0
    if d < 0:
        return None
    
    return a*math.sqrt(d)
```

## Задание 1

```python
a = 5
step = (2 - -2) / 20
x = np.arange(-2, 2 + step, step)
y = [r(a, i) for i in x]

fig = plt.figure(num=1, facecolor='w', edgecolor='k')
fig.set_figwidth(5)
fig.set_figheight(5)

ax = fig.add_subplot(1, 1, 1)
ax.plot(x, y, color='red', label='график',marker='o', linestyle='dotted')
ax.set_title('Вариант 5')
ax.set_xlabel('ось x')
ax.set_ylabel('ось y')
ax.grid()
ax.legend()

plt.show()
```

## Задание 5

```python
def z(x, y):
    return 4 * np.sin(math.pi*2) * np.cos(math.pi*1.5*y)*(1 - x**2) * y * (1 - y) 

step = 0.05
xx = np.arange(-1, 1 + step, step)
yy = np.arange(-1, 1 + step, step)

zz = [z(x,y) for x, y in zip(xx, yy)]

X, Y = np.meshgrid(xx, yy)

Z = z(X, Y)
print(Z)
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.plot(xx, yy, zz, color ='green')
ax.set_title('Вариант 5');
```

## Задание 6

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from matplotlib import cbook, cm
from matplotlib.colors import LightSource

fig = plt.figure(num=1, facecolor='w', edgecolor='k')
fig.set_figwidth(15)
fig.set_figheight(10)
subplots_adjust(hspace=0.3, wspace=0.3)

ax = fig.add_subplot(2, 3, 1, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter('{x:.02f}')
ax.set_title('With LinearLocator')

ax = fig.add_subplot(2, 3, 2, projection='3d')
ax.plot_wireframe(X, Y, Z, color ='green')
ax.set_title('plot_wireframe')

ax = fig.add_subplot(2, 3, 3)
ax.contourf(X, Y, Z, color ='green')
ax.set_title('contourf')

ax = fig.add_subplot(2, 3, 4)
levels = np.linspace(np.min(Z), np.max(Z), 7)
ax.contour(X, Y, Z, levels=levels)
ax.set_title('contour')

# ax = fig.add_subplot(2, 3, 5, projection='3d')
# ax.plot_wireframe(X, Y, Z, color ='green')
# ax.set_title('Вариант 5')

ax = fig.add_subplot(2, 3, 5, projection='3d')
ls = LightSource(270, 45)
rgb = ls.shade(Z, cmap=cm.gist_earth, vert_exag=0.1, blend_mode='soft')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, facecolors=rgb,
                       linewidth=0, antialiased=False, shade=False)

ax.set_title('With LightSource');
```
